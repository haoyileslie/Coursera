```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
file_path = r"C:\Users\lluo\OneDrive - cooperinvestors.com\My Documents\other docs\GTI\FinancialServiceFintech\5206037_Industry_Gross_Value_Added_Annual.xlsx"
df_VGA = pd.read_excel(file_path, sheet_name='Data1')

# Set the index of df to the first column
df_VGA.set_index(df_VGA.columns[0], inplace=True)
df_VGA.tail()
```

```python
# Parse '2023-06-01' into a datetime object
date = pd.to_datetime('2023-06-01')
# Keep only the rows where the index is '2023-06-01'
df_VGA_2023 = df_VGA.loc[df_VGA.index == date]
df_VGA_2023
```

```python
# Split the column names on ';' and keep the first part
df_VGA_2023.columns = df_VGA_2023.columns.str.split(';').str[0]
df_VGA_2023
```

```python
# Remove the last four columns from df
df_VGA_2023 = df_VGA_2023.iloc[:, :-4]
df_VGA_2023
```

```python
# Sum the columns if their names are identical
df_VGA_2023_industry = df_VGA_2023.T.groupby(level=0).sum().T
df_VGA_2023_industry
```

```python
# Sum up all the rows to get a total for each column
totals = df_VGA_2023_industry.sum()

# Sort the totals from high to low
totals_sorted = totals.sort_values(ascending=False)

totals_sorted
```

```python
# Sum up all the rows to get a total for each column
totals = df_VGA_2023_industry.sum()

# Create a pie chart
patches, texts, autotexts = plt.pie(totals, labels=totals.index, autopct='%1.1f%%', pctdistance=1.1, labeldistance=1.2)

# Highlight the specific label and percentage
highlight = 'Financial and insurance services (K)'
for i in range(len(texts)):
    if texts[i].get_text() == highlight:
        texts[i].set_fontsize(14)
        texts[i].set_color('red')
        autotexts[i].set_fontsize(14)
        autotexts[i].set_color('red')
        texts[i].set_bbox(dict(facecolor='yellow', alpha=0.5, edgecolor='black'))
    else:
        texts[i].set_fontsize(8)  # Reduce the font size to 10
        autotexts[i].set_fontsize(8)  # Reduce the font size to 10

# Draw white circles at the center
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

plt.show()
```

```python
# Load the Excel file
file_path = r"C:\Users\lluo\OneDrive - cooperinvestors.com\My Documents\other docs\GTI\FinancialServiceFintech\5206037_Industry_Gross_Value_Added_Annual.xlsx"
df_VGA2 = pd.read_excel(file_path, sheet_name='Data1')

# Set the index of df to the first column
df_VGA2.set_index(df_VGA2.columns[0], inplace=True)
df_VGA2
```

```python
# Ensure that the DataFrame's index only contains valid date strings
df_VGA2.index = pd.to_datetime(df_VGA2.index, errors='coerce')

# Drop any rows with invalid dates
df_VGA2 = df_VGA2.loc[df_VGA2.index.notnull()]

# Parse the dates into datetime objects
start_date = pd.to_datetime('1995-06-01')
end_date = pd.to_datetime('2023-06-01')

# Keep only the rows where the index is between '1995-06-01' and '2023-06-01'
df_VGA2_1995_2023 = df_VGA2.loc[(df_VGA2.index >= start_date) & (df_VGA2.index <= end_date)]
df_VGA2_1995_2023.head()
```

```python
# Split the column names on ';' and keep the first part
df_VGA2_1995_2023.columns = df_VGA2_1995_2023.columns.str.split(';').str[0]
df_VGA2_1995_2023.head()
```

```python
# Sum the columns if their names are identical
df_VGA2_1995_2023_industry = df_VGA2_1995_2023.T.groupby(level=0).sum().T

# Remove the last four columns from df
df_VGA2_1995_2023_industry = df_VGA2_1995_2023_industry.iloc[:, :-4]
df_VGA2_1995_2023_industry.head()
```

```python
# Calculate the number of years
num_years = df_VGA2_1995_2023_industry.index.max().year - df_VGA2_1995_2023_industry.index.min().year

# Calculate the CAGR for each column
cagr = (df_VGA2_1995_2023_industry.iloc[-1] / df_VGA2_1995_2023_industry.iloc[0]) ** (1 / num_years) - 1
cagr
```

```python
# Sort the CAGR values from high to low
cagr_sorted = cagr.sort_values(ascending=True)

# Create a list of colors, where 'red' corresponds to 'Financial and insurance services (K)' and 'blue' corresponds to all other columns
colors = ['red' if 'Financial and insurance services (K)' in index else 'blue' for index in cagr_sorted.index]

# Create a horizontal bar chart of the CAGR values
plt.figure(figsize=(10, 6))
bars = plt.barh(cagr_sorted.index, cagr_sorted.values, color=colors)
plt.xlabel('Compound annual growth rate (CAGR) ')

# # Add the value of each bar at the tip of the bar
# for bar in bars:
#     plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, 
#              '{:.2f}'.format(bar.get_width()), 
#              va='center', ha='left')

plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

# Create a horizontal bar chart of the CAGR values
plt.figure(figsize=(10, 6))
bars = plt.barh(cagr_sorted.index, cagr_sorted.values, color=colors)

# Format the x-axis labels as percentages
formatter = FuncFormatter(lambda y, _: '{:.1%}'.format(y))
plt.gca().xaxis.set_major_formatter(formatter)

plt.xlabel('Compound annual growth rate (CAGR)')

# Add the value of each bar at the tip of the bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, 
             '{:.1f}%'.format(bar.get_width() * 100), 
             va='center', ha='left')

plt.show()
```

```python
import pandas as pd

# Sort the DataFrame in descending order
df_FDI_industry_2022_sorted = df_FDI_industry_2022.sort_values(ascending=False)

# Create two separate lists: one for the first half of the sorted series and one for the second half
first_half = df_FDI_industry_2022_sorted.index[:len(df_FDI_industry_2022_sorted)//2]
second_half = df_FDI_industry_2022_sorted.index[-(len(df_FDI_industry_2022_sorted)//2):][::-1]

# Interleave the two lists
zigzag_indices = [index for pair in zip(first_half, second_half) for index in pair]
if len(df_FDI_industry_2022_sorted) % 2:
    zigzag_indices.append(df_FDI_industry_2022_sorted.index[len(df_FDI_industry_2022_sorted)//2])
df_FDI_industry_2022_zigzag = df_FDI_industry_2022_sorted.loc[zigzag_indices]

# Create a pie chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.pie(df_FDI_industry_2022_zigzag, labels = 
       [f'{index.split(";")[0]}: ${value/1000:.1f}bn, {value/df_FDI_industry_2022_zigzag.sum()*100:.1f}%' for index, value in df_FDI_industry_2022_zigzag.items()], 
       textprops={'fontsize': 8})

# Display the plot
plt.show()
```

```python
import pandas as pd

# Load the file
df_gdp = pd.read_csv(r'C:\Users\lluo\OneDrive - cooperinvestors.com\My Documents\other docs\GTI\FinancialServiceFintech\OECDoutlookAU.csv')

# Keep only the rows where 'Country' is 'Australia' and 'Variable' is 'Gross domestic product, volume, market prices'
df_gdp_au = df_gdp.loc[(df_gdp['Country'] == 'Australia') & (df_gdp['Variable'] == 'Gross domestic product, volume, market prices')]

df_gdp_au.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Filter the DataFrame to include only the years from 2014 onwards
df_gdp_au = df_gdp_au[df_gdp_au['Time'] >= 2013]

# Convert the values into trillion
df_gdp_au['Value'] = df_gdp_au['Value'] / 1e12

# Create a list of colors
colors = ['red' if year in [2024, 2025] else 'blue' for year in df_gdp_au['Time']]

# Plot a bar chart
plt.figure(figsize=(15, 8))
bars = plt.bar(df_gdp_au['Time'], df_gdp_au['Value'], color=colors)
plt.xlabel('Year')
plt.ylabel('Gross domestic product, volume, market prices in AU$ (trillion)')

# Set the start and end year on the x-axis
plt.xlim(2012, 2026)

# Set the y-axis to fit the largest value of the bars
plt.ylim(0, df_gdp_au['Value'].max() * 1.1)  # Add a 10% margin

# Label each year on the x-axis
plt.xticks(df_gdp_au['Time'])

# Format the y-axis labels to hide the scientific notation
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'{x:.2f}T'))

# Add the values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```
